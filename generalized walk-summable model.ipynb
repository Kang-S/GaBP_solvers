```python
import numpy as np
import matplotlib.pyplot as  plt
import solvers

%matplotlib inline
```

```python
A = np.array([[10, 1.5, 2], [2, 4, 2.5], [2, 3, 5]])
B = np.array([[10, 0.5], [0.5, 5]])
C = np.array([[7, 1], [1, 2]])
A_B = np.array([[2, 0], [0, 2], [0, 0]])
A_C = np.array([[2, 0], [0, 0], [0, 1]])
B_C = np.array([[1, 0], [0, 1]])
rhs = np.random.randn(7)
D = np.block([[A, A_B, A_C], [A_B.T, B, B_C], [A_C.T, B_C.T, C]])
np.linalg.eig(D)[0]
```

```python
D
```

GaBP

```python
DD = abs(np.eye(7) - np.diag(1/np.diag(D))@D)
np.linalg.eig(DD)[0]
```

```python
x = solvers.GaBP(D, rhs, verbose=True)
```

Generalized GaBP

```python
A_B_ = abs(np.linalg.inv(A)@A_B)
A_C_ = abs(np.linalg.inv(A)@A_C)
B_A_ = abs(np.linalg.inv(B)@A_B.T)
B_C_ = abs(np.linalg.inv(B)@B_C)
C_A_ = abs(np.linalg.inv(C)@A_C.T)
C_B_ = abs(np.linalg.inv(C)@B_C.T)
```

$l_1$ norm

```python
a_b = np.sum(A_B_.reshape(-1,))
a_c = np.sum(A_C_.reshape(-1,))
b_a = np.sum(B_A_.reshape(-1,))
b_c = np.sum(B_C_.reshape(-1,))
c_a = np.sum(C_A_.reshape(-1,))
c_b = np.sum(C_B_.reshape(-1,))
d = np.array([[0, a_b, a_c], [b_a, 0, b_c], [c_a, c_b, 0]])
np.linalg.eig(d)[0]
```

Frobenius norm

```python
a_b = np.linalg.norm(A_B_.reshape(-1,))
a_c = np.linalg.norm(A_C_.reshape(-1,))
b_a = np.linalg.norm(B_A_.reshape(-1,))
b_c = np.linalg.norm(B_C_.reshape(-1,))
c_a = np.linalg.norm(C_A_.reshape(-1,))
c_b = np.linalg.norm(C_B_.reshape(-1,))
d = np.array([[0, a_b, a_c], [b_a, 0, b_c], [c_a, c_b, 0]])
np.linalg.eig(d)[0]
```

$l_{\infty}$ norm

```python
a_b = np.max(A_B_.reshape(-1,))
a_c = np.max(A_C_.reshape(-1,))
b_a = np.max(B_A_.reshape(-1,))
b_c = np.max(B_C_.reshape(-1,))
c_a = np.max(C_A_.reshape(-1,))
c_b = np.max(C_B_.reshape(-1,))
d = np.array([[0, a_b, a_c], [b_a, 0, b_c], [c_a, c_b, 0]])
np.linalg.eig(d)[0]
```

Maximum row

```python
a_b = np.max(np.sum(A_B_.T, axis=0))
a_c = np.max(np.sum(A_C_.T, axis=0))
b_a = np.max(np.sum(B_A_.T, axis=0))
b_c = np.max(np.sum(B_C_.T, axis=0))
c_a = np.max(np.sum(C_A_.T, axis=0))
c_b = np.max(np.sum(C_B_.T, axis=0))
d = np.array([[0, a_b, a_c], [b_a, 0, b_c], [c_a, c_b, 0]])
np.linalg.eig(d)[0]
```

```python
a_b = np.sum(np.max(np.linalg.svd(A_B_)[1]))
a_c = np.sum(np.max(np.linalg.svd(A_C_)[1]))
b_a = np.sum(np.max(np.linalg.svd(B_A_)[1]))
b_c = np.sum(np.max(np.linalg.svd(B_C_)[1]))
c_a = np.sum(np.max(np.linalg.svd(C_A_)[1]))
c_b = np.sum(np.max(np.linalg.svd(C_B_)[1]))
d = np.array([[0, a_b, a_c], [b_a, 0, b_c], [c_a, c_b, 0]])
np.linalg.eig(d)[0]
```
